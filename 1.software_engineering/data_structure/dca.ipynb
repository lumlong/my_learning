```python
# Heap data structure

def heapify(arr, n, i):
    '''
    This takes
    array :
    size : size of array
    i : index of array
    '''
    
    # set i as largest
    largest = i
    # pos of left node is 2i + 1
    l = 2 * i + 1
    # pos of rightt node is 2i + 2
    r = 2 * i + 2 
    
    # l pos must still in size, compare if i < l
    if l < n and arr[i] < arr[l]:
        # set largest as l
        largest = l
    
    if r < n and arr[largest] < arr[r]:
        largest = r
    
    # if new largest
    if largest != i:
        # swap
        arr[i],arr[largest] = arr[largest],arr[i]
        # recursive heapify until i is largest
        heapify(arr, n, largest)
        

# 
def insert(array, newNum):
    
    size = len(array)
    
    # if heap is empty
    if size == 0:
        array.append(newNum)
        
    else:
        array.append(newNum)
        # Loop all element with heapify from last to the root
        for i in range((size//2)-1, -1, -1):
            heapify(array, size, i)

            
            
def deleteNode(array, num):
    size = len(array)
    i = 0
    for i in range(0, size):
        if num == array[i]:
            break
        
    array[i], array[size-1] = array[size-1], array[i]

    array.remove(num)
    
    for i in range((len(array)//2)-1, -1, -1):
        heapify(array, len(array), i)


arr = []

insert(arr, 3)
insert(arr, 4)
insert(arr, 9)
insert(arr, 5)
insert(arr, 2)

print ("Max-Heap array: " + str(arr))

deleteNode(arr, 4)
print("After deleting an element: " + str(arr))
```

```python
import math


class Node():
    value = None
    childs = []
    degree = 0
    
    def __init__(self, value:int):
        self.value = value
    
    
class FibonacciHeap():
    # Set of heap-degreeed trees.
    rootNodes = []
    # Maintain pointer to minimum element.
    minNode = None
    nodeCount = 0
    
    def __init__(self):
        pass
    
    def insert(self, value):
        newNode = Node(value)
        self.rootNodes.append(newNode)
        # set new min pointer
        if (self.minNode is None or value < self.minNode.value):
            self.minNode = newNode
        self.nodeCount += 1
        
        
    def _linking(self, nodeA:Node, nodeB:Node):
        """
        Make larger root node be a child of smaller root node.
        """
        x = nodeA
        y = nodeB
        if x > y:
            x,y = y,x
        x.childs.append(y)
        x.degree += 1
        return x
            
    
    def _floor_log(x):
        return math.frexp(x)[1] - 1

    def _consolidate(self):
        # Create empty list size of maxDegree of tree in heap
        maxDegree = floor_log(self.nodeCount) + 1
        degreeList = maxDegree * [None]
        
        # Iterate thought rootNodes, empty it to degreeList,
        while self.rootNodes != []:
            x = self.rootNodes[0]
            degree = x.degree
            self.rootNodes.remove(x)
            
            # Check the degree of currentNode with the degreeList
            # If the degree in degreeList is not empty
            while degreeList[degree] is not None:
                y = degreeList[degree]
                # Linking the nodes
                if x.value > y.value:
                    x,y = y,x
                x.childs.append(y)
                x.degree += 1
                # Remove the linked node from the currentNodeDegree index
                degreeList[degree] = None
                # Update degree to the next one
                degree += 1
            # If the degree is empty, assign new one to it
            degreeList[degree] = x

        # Update min
        self.minNode = None
        # Reconstruct rootList
        for currentNode in degreeList:
            if currentNode is not None:
                self.rootNodes.append(currentNode)
                # Check for min
                if (self.minNode is None
                    or currentNode.value < self.minNode.value):
                    self.minNode = currentNode
    
                    
    def extract_min(self):
        
        minNode = self.minNode
        
        if minNode is not None:
            # melt childs of minNode to rootNode
            for child in self.minNode.childs :
                self.rootNodes.append(child)

            # delete minNode from rootNodes
            self.rootNodes.remove(minNode)
            if self.rootNodes == [] :
                self.minNode = None
            else :
                self.minNode = self.rootNodes[0]
                self._consolidate()
            self.nodeCount = self.nodeCount - 1
            return minNode.value
```
